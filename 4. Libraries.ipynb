# 4.3 Pandas DataFrame
Series를 모아 동일한 인덱스로 묶어주는 데이터프레임을 생성

```python
# Import libraries
import pandas as pd

# 첫 번째 변수는 단순한 리스트
# index 라는 변수에 할당된 리스트
# 서울 인구 100, 부산 80, 대구 60, 광주 50 등
a = pd.Series([100, 80, 60, 50], index=['서울', '부산', '대구', '광주'])

print(a, '\n')
print('a[0] = ', a['서울'])
print("a['서울'] = ", a['서울'])

a.name = '인구'
a.index.name = '도시'
print('\n', a)
# dataframe : Series들을 모아놓은 것
# 데이터프레임의 각 행에 인덱스가 정의될 수 있음
# 시리즈는 1차원 리스트, 데이터 프레임은 2차원 배열(인덱스를 곁들인)

# dictionary는 key와 value로 구성
a = {
    'name': ['kim', 'lee', 'park', 'choi'],  # name 에 대한 value list
    'grade': ['A', 'B', 'C', 'D']  # grade 에 대한 value 리스트
}
index = ['001', '002', '003', '004']

# 데이터프레임 생성 (dictionary와 index를 제공)
# 인덱스를 별도로 지정해줄 수 있음
df = pd.DataFrame(a, index=index)

# 데이터프레임 생성 (dictionary만 제공)
# 인덱스를 지정하지 않았기 때문에 0, 1, 2, .. 의 기본 index가 생성됨
df1 = pd.DataFrame(a)

print(df, '\n')
print(df1, '\n')

a = {'name': ['kim', 'lee', 'park', 'choi'], 'grade': ['A', 'B', 'C', 'D']}
index = ['001', '002', '003', '004']

print('특정 열만 선택 : ')
print(df['grade'], '\n')

print('특정 열과 특정 행 선택 : ')
print(df['grade']['001'], '\n')
print(df.values[0][1], '\n')
print(df.values[0][0], '\n')
```

# 4.4 Matplotlib - 영상 출력

```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread('data/lena.jpg')
img2 = cv2.imread('data/chessboard.jpg')
img3 = cv2.imread('data/sample.jpg')

plt.figure(figsize=(5,5)) # 영상을 그릴 화면, figsize : 전체 그림이 나타날 크기
plt.subplot(2, 2, 1) # 영상이 표시될 subplot, 2, 2, 1 위치에 출력

# 색상의 split
b, g, r = cv2.split(img1)
img1 = cv2.merge([r, g, b])

b, g, r = cv2.split(img2)
img2 = cv2.merge([r, g, b])

plt.imshow(img1)
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2)
plt.imshow(img2)
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3)
plt.imshow(img3)
plt.xticks([]), plt.yticks([])

plt.show()
```
